```python
%pylab inline
%debug
from __future__ import division
import numpy as np
from numba import *

#@jit(restype=None, argtypes=[int32], locals={'k':int32,'I':object_,'Q':object_,'symbArray':object_,'Esnorm':int32, 'Ebnorm':int32, 'Es':int32, 'Eb':int32}, backend='ast', target='cpu', nopython=False)

def modulationScheme(M):
    # -----------------
    k = np.log2(M)                              # Number of bits per symbol
    if M == 2:
        I = np.array([-1,1], dtype = np.int32)
        Q = 1j *np.array([0,0], dtype = np.int32)
        symbArray = I + Q
        Esnorm = np.mean(np.abs(symbArray)**2)         # Energy per symbol, the sum(Sr.^2+Si.^2)/(Number of symbols)
        Ebnorm = Esnorm/k
        symbArray = symbArray/np.sqrt(Esnorm)       # Normalised symbol array
        
        Es = 1                                      # normalisation
        Eb = Es/k                                   # Energy per bit
        return k,Es,Esnorm,Eb,Ebnorm,symbArray
    elif M == 4:
        I = np.array([-1,1,-1,1], dtype = np.int32)
        Q = 1j * np.array([-1,-1,1,1], dtype = np.int32)
        symbArray = I + Q
        Esnorm = np.mean(np.abs(symbArray)**2)         # Energy per symbol, the sum(Sr.^2+Si.^2)/(Number of symbols)
        Ebnorm = Esnorm/k
        symbArray = symbArray/np.sqrt(Esnorm)       # Normalised symbol array
        
        Es = 1                                      # normalisation
        Eb = Es/k                                   # Energy per bit
        return k,Es,Esnorm,Eb,Ebnorm,symbArray
    elif M == 16:
        #The real and imaginary axes are often called the in phase, or I-axis, and the quadrature, or Q-axis, respectively
        I = np.array([-3,-3,-3,-3,-1,-1,-1,-1,3,3,3,3,1,1,1,1], dtype = np.int32)
        Q = 1j * np.array([3,1,-3,-1,3,1,-3,-1,3,1,-3,-1,3,1,-3,-1],dtype = np.int32)
        symbArray = I + Q
        Esnorm = np.mean(np.abs(symbArray)**2)         # Energy per symbol, the sum(Sr.^2+Si.^2)/(Number of symbols)
        Ebnorm = Esnorm/k
        symbArray = symbArray/np.sqrt(Esnorm)       # Normalised symbol array
        
        Es = 1                                      # normalisation
        Eb = Es/k                                   # Energy per bit
        return k,Es,Esnorm,Eb,Ebnorm,symbArray
    
def amplifierModel( MUx,backoff ):
    # PA Characteristics
    AlphaA = 1
    BetaA = 1
    AlphaPhi = np.pi/48
    BetaPhi = 0.5
    p=3
    n = len(MUx)
    y = np.zeros((n),dtype = np.complex) # initialize output array
    amam = np.zeros((n), dtype = np.float)
    ampm = np.zeros((n), dtype = np.float)
    phase = np.zeros((n), dtype = np.float)
    
    absMUx = np.abs(MUx)
    phaseMUx = np.angle(MUx)
    Bk = 10**(backoff/20)
    for k in np.arange(n):
        x = Bk*absMUx[k]
        amam[k] = (AlphaA*x)/(1 + (BetaA*x)**2*p)**(1/(2*p))
        ampm[k] = (AlphaPhi*x**2)/(1+BetaPhi*x**2)  #change in Phase
        phase[k] = phaseMUx[k]+ ampm[k]
        y[k] = amam[k]*np.exp(1j*phase[k])
    return amam, ampm, y


bandWidth = 5e6
occupiedBW = 4.5e6
frequencySpacing = 15e3
symbolDuration = 1/frequencySpacing
NFFT = 512
Nsubcarriers = 300
timeDomainSampleRate = symbolDuration/NFFT
acctualBW = 1/timeDomainSampleRate
overSampleRate = NFFT/Nsubcarriers
frameDuration = 1e-2
samplesPerFrame = frameDuration/timeDomainSampleRate
Nsubframes = 10
samplesPerSubframe = samplesPerFrame/Nsubframes
subframeDuration = samplesPerSubframe * timeDomainSampleRate
NslotsperSubframe = 2
samplesPerSlot = samplesPerSubframe/NslotsperSubframe
slotDuration = subframeDuration/NslotsperSubframe

print "Script Configuration"
print "-------------------------------------------------------\n"
print("Channel Bandwidth             : {}".format(bandWidth))
print("Occupied Bandwidth            : {}".format(occupiedBW))
print("Frequency Spacing             : {}".format(frequencySpacing))
print("Symbol Duration               : {}".format(symbolDuration))
print("Number of Frequency Points    : {}".format(NFFT))
print("Number of Subcarriers         : {}".format(Nsubcarriers))
print("Time Domain Sample Rate       : {}".format(timeDomainSampleRate)) 
print("Actual Bandwidth              : {}".format(acctualBW)) # SampleRate
print("Over Sample Rate              : {}".format(overSampleRate))
print("Frame Duration                : {}".format(frameDuration))
print("Samples Per Frame             : {}".format(samplesPerFrame))
print("Number of Subframes per Frame : {}".format(Nsubframes))
print("Samples Per Subframe          : {}".format(samplesPerSubframe))
print("Subframe Duration             : {}".format(subframeDuration))
print("Slots Per subFrame            : {}".format(NslotsperSubframe))
print("Samples Per Slot              : {}".format(samplesPerSlot))
print("Subframe Duration             : {}".format(slotDuration))
print "-------------------------------------------------------\n"

slotConfig = np.array([40,	512,	36,	512,	36,	512,	36,	512,	36,	512,	36,	512,	36,	512], dtype = np.int32)
print "Slot Configuration"
print "-------------------------------------------------------\n"
print("Slot Config:\n{}".format(slotConfig))
print "-------------------------------------------------------\n"
slotConfig = np.reshape(slotConfig,(-1,2))
NumberofSym = slotConfig.shape[0]

M = 4

SubCarrierIndex = np.concatenate((np.arange(0,Nsubcarriers/2,dtype=np.int32),np.arange((NFFT-1) - Nsubcarriers/2,NFFT-1,dtype=np.int32)),axis = 0)
fs = occupiedBW/(Nsubcarriers-1)
frequencyRange = np.arange(0,occupiedBW+1,fs)
Nrb = 25
k,Es,Esnorm,Eb,Ebnorm,symbArray = modulationScheme(M)

NumberOfOFDMPerSub = NslotsperSubframe * NumberofSym
NumberOfOFDMPerFrame = NumberOfOFDMPerSub * Nsubframes

print("Number of OFDM Symbols per Frame: {}".format(NumberOfOFDMPerFrame))
print("Number of OFDM Symbols per Sub Frame: {}".format(NumberOfOFDMPerSub))
print("Number of OFDM Symbols per Slot: {}".format(NumberofSym))
print("Number of bits Per Symbol: {}".format(k))
print("Energy Per Symbol: {}".format(Es))
print("Energy Per Bit: {}".format(Eb))

print ""
print "Normalised Symbol Array"
print "-------------------------------------------------------"
print("Normalised Symbol Array:\n{}".format(symbArray))
print "-------------------------------------------------------\n"

print "Subcarrier Index"
print "-------------------------------------------------------"
print("Subcarrier Index: {0} to {1}".format(SubCarrierIndex[0], SubCarrierIndex[-1]))
print("Sample Frequency: {}".format(fs))
print("Frequency Range: {0} to {1}".format(frequencyRange[0], frequencyRange[-1]))
print("Number of Resource Blocks:{}".format(Nrb))
print "-------------------------------------------------------\n"

subplot(2,1,1)
stem(SubCarrierIndex,frequencyRange)
grid(True)
xlabel("Subcarrier Index")
ylabel("Frequency")
title("Subcarrier Index vs. Frequency")
subplot(2,1,2)
stem(SubCarrierIndex,fftshift(frequencyRange))
grid(True)
xlabel("Subcarrier Index")
ylabel("Frequency")

nFrames = 10
nSubFrames = nFrames * Nsubframes
nSlots = nSubFrames * NslotsperSubframe
nOFDM = nSlots * NumberofSym
nSymb = Nsubcarriers * NumberOfOFDMPerFrame * nFrames
nBits = nSymb * k
print "Transmition"
print "-------------------------------------------------------"
print("Number of Frames to be Transmitted: {}".format(nFrames))
print("Number of Sub Frames to be Transmitted: {}".format(nSubFrames))
print("Number of Slots to be Transmitted: {}".format(nSlots))
print("Number of OFDM Symbols to be Transmitted: {}".format(nOFDM))
print("Number of Symbols to be Transmitted: {}".format(nSymb))
print("Number of Bits to be Transmitted: {}".format(nBits))
print "-------------------------------------------------------"


Bits = np.array(np.random.rand(1,nBits) > 0.5,dtype = np.int32) 
Bits = np.reshape(Bits,(-1,k,nOFDM))
```
