```python
from __future__ import division
import numpy as np
from numba import *

class pathlossmodels():
    def __init__(self, fMHz, Henb, Hue, Dkm):
        self.fMHz = fMHz
        self.Henb = Henb
        self.Hue = Hue
        self.Dkm = Dkm
        
    def gdHataModel(self,environment):
        fMHz = self.fMHz
        Henb = self.Henb
        Hue = self.Hue
        Dkm = self.Dkm
        
        if environment == 'Urban':
            K = 2 * (np.log10(fMHz/28))**2 + 5.4 # Suburban
            Cm = 3
        elif environment == 'Suburban':
            K = 2 * (np.log10(fMHz/28))**2 + 5.4 # Suburban
            Cm = 0
        elif environment == 'Rural':
            K = 4.78*(np.log10(fMHz))**2 - 18.33*np.log10(fMHz) + 40.94 #Open
            Cm = -3
            
        cost231GD = 46.33 + (44.9 -6.55*np.log10(Henb))*np.log10(Dkm) + 33.9*np.log10(fMHz) - self.a(Hue) - 13.82*np.log10(Henb) -K + Cm
        return cost231GD
    
    def cost231(self,environment):
        fMHz = self.fMHz
        Henb = self.Henb
        Hue = self.Hue
        Dkm = self.Dkm
        
        envOptions = {'Urban'  : 3,
                     'Suburban' : 0,
                     'Rural' : -3}
        Cm = envOptions[environment]
        cost231 = 46.33 + (44.9 -6.55*np.log10(Henb))*np.log10(Dkm) + 33.9*np.log10(fMHz) - self.a(Hue) - 13.82*np.log10(Henb) + Cm
        return cost231
    
    def a(self,Hue):
        return (1.1*np.log10(self.fMHz) - 0.7)*Hue - (1.56 * np.log10(self.fMHz) - 0.8)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)



fMHz = 700
Henb = 30
Hue = 1.5
Dkm = np.arange(30)
PL = pathlossmodels(fMHz, Henb, Hue, Dkm)
ax.plot(xs=Dkm, ys=np.ones(len(Dkm))*fMHz, zs=PL.gdHataModel('Urban'), zdir='z', label='ys=0, zdir=z')
hold(True)
ax.plot(xs=Dkm, ys=np.ones(len(Dkm))*fMHz, zs=PL.gdHataModel('Suburban'), zdir='z', label='ys=0, zdir=z')
ax.plot(xs=Dkm, ys=np.ones(len(Dkm))*fMHz, zs=PL.gdHataModel('Rural'), zdir='z', label='ys=0, zdir=z')
ax.plot(xs=Dkm, ys=np.ones(len(Dkm))*fMHz, zs=PL.cost231('Urban'), zdir='z', label='ys=0, zdir=z')
ax.plot(xs=Dkm, ys=np.ones(len(Dkm))*fMHz, zs=PL.cost231('Suburban'), zdir='z', label='ys=0, zdir=z')
ax.plot(xs=Dkm, ys=np.ones(len(Dkm))*fMHz, zs=PL.cost231('Rural'), zdir='z', label='ys=0, zdir=z')
ax.set_xlabel('Distance')
ax.set_ylabel('Frequency')
ax.set_zlabel('Path Loss')
legend(['GD Urban','GD Suburban', 'GD Rural','COST231 Urban','COST231 Suburban', 'COST231 Rural'],loc ='lower left',prop={'size':8},shadow = True, fancybox = True)
#legend(['GD Urban','GD Suburban', 'GD Rural','COST231 Urban','COST231 Suburban', 'COST231 Rural'])
#show()
figure()
fMHz = np.arange(700,2500,100)
Dkm = 15
PL = pathlossmodels(fMHz, Henb, Hue, Dkm)
plot(fMHz,PL.gdHataModel('Rural') )
xlabel('Frequency')
ylabel('Path Loss')
grid(True)

```
