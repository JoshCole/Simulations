```python
%%cython
from __future__ import division
import numpy as np
cimport numpy as np #overwritten those from python with cython
import cython
cimport cython
from scipy.special import erfc

ctypedef np.int32_t DTYPEint_t
ctypedef np.float64_t DTYPEfloat_t
ctypedef np.complex128_t DTYPEcomplex_t

@cython.boundscheck(False)
@cython.wraparound(False)
@cython.nonecheck(False)
def cyofdmSim(np.int32_t nSym):
    cdef DTYPEint_t N, Nsp, Nfft, ofdmBW, ebMin, ebMax, ebStep, ebRange
    # Simulation Paramters
    ebMin = -20
    ebMax = 8
    ebStep = 2
    #ebRange = np.divide((ebMax - ebMin), ebStep)
    #print np.ceil((ebMax - ebMin)/ebStep)
    ebRange = 14 #np.int(np.ceil((ebMax - ebMin)/ebStep))
    
    cdef np.ndarray[DTYPEint_t,ndim=1] EbNodB 

    EbNodB = np.arange(ebMin,ebMax,ebStep,dtype = np.int32) # Bit to noise ratio
    
    #--------------------------

    # OFDM Parameters - Giver in IEEE Spec
    
    N = 64 # FFT size or total number of subcarriers (used + unused) 64
    Nsd = 48 # Number of data subcarriers 58
    Nsp = 4 # number of pilot subcarriers 4
    ofdmBW = 20 * 10**6
    #--------------------------
    
    cdef DTYPEint_t Ncp, Nst, nBitsperSym,numErrors
    cdef DTYPEfloat_t deltaF, Tfft, Tgi, Tsignal, theoreticalBER
    cdef Py_ssize_t i, j

    # Derived Parameters
    deltaF = ofdmBW/N # = 20 MHz/64 = 0.3125 MHz
    Tfft = 1/deltaF # IFFT/FFT period = 3.2us
    Tgi = Tfft/4 # Guard interval duration - duration of cyclic prefix
    Tsignal = Tgi + Tfft # duration of BPSK-OFDM symbol
    Ncp = np.divide(N*Tgi,Tfft) # Number of symbols allocated to cyclic prefix
    Nst = Nsd + Nsp # Number of total used subcarriers
    nBitsPerSym =Nst # For BPSK the number of Bits per Symbol is same as num of sub carriers
    Nfft = N+Ncp
    
    cdef np.ndarray[DTYPEint_t,ndim=1] errors = np.zeros(ebRange, dtype = np.int32), s
    cdef np.ndarray EsNodB, X_Freq, x_Time

    #simulatedBER = np.zeros((ebRange), dtype = np.float64)
  
    EsNodB = EbNodB + 10*np.log10(Nst/N) + 10*np.log10(N/(Ncp + N)) # converting to symbols to noise ratio
    
    cdef np.ndarray noise, r, r_Parallel, r_Time
    cdef np.ndarray[DTYPEcomplex_t,ndim=1] R_Freq = np.zeros(N, dtype = np.complex128), s_cap = np.zeros(N, dtype = np.complex128)
    
    noise = 1/np.sqrt(2)*(np.random.randn(nSym, Nfft) + 1j*np.random.randn(nSym, Nfft))
    for i in range(ebRange):
        for j in range(nSym):
            
            s = np.int32(2*np.round(np.random.rand(Nst))-1) # Generating Random Data with BPSK modulation
            # IFFT block
            # Assigning subcarriers from 1 to 26 (mapped to 1-26 of IFFT input)
            # and -26 to -1 (mapped to 38 to 63 of IFFT input); Nulls from 27 to 37
            # and at o position
            X_Freq = np.concatenate((np.zeros(1),s[0:Nst/2],np.zeros(11),s[Nst/2::]),axis = 0)
            # Pretending the data to be in frequency domain and converting to time domain
            x_Time = N/np.sqrt(Nst)*np.fft.ifft(X_Freq)
            # Adding Cyclic Prefix
            ofdm_signal = np.concatenate((x_Time[N-Ncp:N],x_Time),axis = 0)
            # Channel Modeling
            
            r = np.sqrt((N + Ncp)/N)*ofdm_signal + 10**(-EsNodB[i]/20)*noise[j,:]
            # Receiver
            # Removing cyclic prefix
            r_Parallel = r[Ncp:(N+Ncp)]
            
            # FFT Block
            r_Time = np.sqrt(Nst)/N*(np.fft.fft(r_Parallel))
            # Extracting the data carriers form the FFT ouput
            
            R_Freq = r_Time[np.concatenate((np.arange(1,Nst/2+1,dtype = np.int32),np.arange(Nst/2+12,Nst+12,dtype = np.int32)),axis = 0)]
               
            # BPSK demodulation / Constelation Demapper
            
            R_Freq[R_Freq>0] = +1
            R_Freq[R_Freq<0] = -1
            s_cap = R_Freq
            numErrors = sum(abs(s_cap-s)/2) # Count number of errors
            # Accumulate bit errors for all symbols transmitted
            errors[i] = errors[i] + numErrors
        
    simulatedBER = errors/(nSym*Nst)
    #theoreticalBER = (1/2)*erfc(np.sqrt(10**(EbNodB/10)))  #Same sa BER for BPSK over AWGN
    return 
```

```python
%%cython

from cython.view cimport array as cvarray
import numpy as np

# Memoryview on a NumPy array
narr = np.arange(27, dtype=np.dtype("i")).reshape((3, 3, 3))
cdef int [:, :, :] narr_view = narr

# Memoryview on a C array
cdef int carr[3][3][3]
cdef int [:, :, :] carr_view = carr

# Memoryview on a Cython array
cyarr = cvarray(shape=(3, 3, 3), itemsize=sizeof(int), format="i")
cdef int [:, :, :] cyarr_view = cyarr

# Show the sum of all the arrays before altering it
print "NumPy sum of the NumPy array before assignments:", narr.sum()

# We can copy the values from one memoryview into another using a single
# statement, by either indexing with ... or (NumPy-style) with a colon.
carr_view[...] = narr_view
cyarr_view[:] = narr_view
# NumPy-style syntax for assigning a single value to all elements.
narr_view[:, :, :] = 3

# Just to distinguish the arrays
carr_view[0, 0, 0] = 100
cyarr_view[0, 0, 0] = 1000

# Assigning into the memoryview on the NumPy array alters the latter
print "NumPy sum of NumPy array after assignments:", narr.sum()

# A function using a memoryview does not usually need the GIL
cpdef int sum3d(int[:, :, :] arr) nogil:
    cdef int total = 0
    I = arr.shape[0]
    J = arr.shape[1]
    K = arr.shape[2]
    for i in range(I):
        for j in range(J):
            for k in range(K):
                total += arr[i, j, k]
    return total

# A function accepting a memoryview knows how to use a NumPy array,
# a C array, a Cython array...
print "Memoryview sum of NumPy array is", sum3d(narr)
print "Memoryview sum of C array is", sum3d(carr)
print "Memoryview sum of Cython array is", sum3d(cyarr)
# ... and of course, a memoryview.
print "Memoryview sum of C memoryview is", sum3d(carr_view)
```
